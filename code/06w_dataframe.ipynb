### 함수 df.query(조건)
- 조건을 만족하는 행 표시

```python
df = pd.DataFrame({'A': range(1, 6),
                   'B': range(6, -3, -2),
                   'C': range(10, 5, -1),
                   'D clmns': range(3, -2, -1),
                   'E': ['python', 'java', 'python', 'go', 'python']})
df
```

```python
df.query('A > B')
```

```python
df[df.A > df.B]
```

```python
df.query('C == 10')
```

```python
df.query('A > 0')
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")
```

##### 열명 중간에 빈 문자열이 있는 경우
- 반드시 열 명을 백쿼트(back quote, Tab키 위의 키) `D clmns`를 사용
- "D clmns"는 오류가 발생하지 않으나 원하는 결과 안나옴

```python
df.query('A == `D clmns`')
```

```python
df.query('A == "A clmns"')
```

```python
df[df.A == df["D clmns"]]
```

```python
df.query('A > 3 | B > 8')
```

```python
df.query('A > 2 & C < 8')
```

```python
df.query("E in ['python', 'go']")
```

#### df.query() 내부의 질의 문자열 내부에서 변수 사용
- @변수명을 사용
- f 문자열로 사용이 가능

```python
num = 3
df.query('B > @num')
```

```python
num = 3
df.query(f'B > {num}')
```

```python
s = "E.str.contains('py')"
df.query(s) # '부분문자열'이 포함된 행의 데이터프레임
```

```python
s = 'E.str.startswith("g")'
df.query(s) # '시작문자열'로 시작되는 행을 반환
```

```python
s = "E.str.endswith('on')"
df.query(s) # '끝나는문자열'로 종료되는 행의 데이터프레임을 반환
```
