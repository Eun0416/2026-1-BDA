### integers()

### standard_norml()

```python
import numpy as np

rg = np.random.default_rng(1)
rg.standard_normal(3) # 표준 정규 분포 실수 난수
```

```python
rg.standard_normal((3, 4)) # 표준 정규 분포 실수 난수
```

## 배열 결합 연습

```python
a = np.arange(1, 4)
b = np.arange(4, 7)
c = np.arange(7, 10)
a, b, c
```

```python
p = np.arange(1, 7).reshape(2, 3)
q = np.arange(11, 17).reshape(2, 3)
r = np.arange(21, 27).reshape(2, 3)
p, q, r
```

```python
np.vstack( (a, b, c) ) # 1D → 2D
```

```python
np.vstack( (p, q, r) ) # 2D → 2D
```

```python
np.hstack( (a, b, c) ) # 1D → 1D
```

```python
np.hstack( (p, q, r) ) # 2D → 2D
```

```python
a, b, c
```

```python
np.column_stack((a, b, c)) # 1D → 2D
```

```python
np.column_stack((p, q, r)) # 2D → 2D
```

```python
np.concatenate((a, b, c))
```

```python
np.concatenate((a, b, c), axis=0)
```

```python
np.concatenate((a, b, c), axis=1) # 1차원에는 축이 존재하지 않음
```

```python
np.concatenate((a, b, c), axis=None)
```

```python
np.concatenate((a, b, c), axis=-1)
```

```python
p, q, r
```

```python
np.concatenate((p, q, r))
```

```python
np.concatenate((p, q, r), axis=0)
```

```python
np.concatenate((p, q, r), axis=1)
```

```python
np.concatenate((p, q, r), axis=-1)
```

```python
np.concatenate((p, q, r), axis=None) # 1차원으로 나열해줌
```

```python
np.c_[a, b, c] # 1D → 2D # 괄호 잘 안씀
```

```python
np.column_stack( (a, b, c) ) # 1D → 2D
```

```python
np.column_stack( a, b, c ) # 리스트나 튜플 형태로 묶어야 함.
```

```python
np.c_[p, q, r] # 2D → 2D
```

```python
np.r_[a, b, c] # 1D → 1D
```

```python
np.concatenate((a, b, c))
```

```python
np.r_[p, q, r]
```

```python
np.vstack((p, q, r))
```
